With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```
data = pd.read_csv('crop_yield.csv')
```

```
data
```

```
data.isna().sum()
```

```
sns.scatterplot(x = data['Annual_Rainfall'], y = data['Yield'])
plt.show()
```

```
year_yield = data.groupby('Crop_Year').sum()
year_yield
```

```
plt.figure(figsize = (12,5))
plt.plot(year_yield.index, year_yield['Yield'],color='blue', linestyle='dashed', marker='o',
        markersize=12, markerfacecolor='yellow')
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Measure of Yield over the year')
plt.show()
```

```
plt.figure(figsize = (12,10))
plt.plot(year_yield.index, year_yield['Area'],color='blue', linestyle='dashed', marker='o',
        markersize=12, markerfacecolor='red')
plt.xlabel('Year')
plt.ylabel('Area')
plt.title('Area under cultivation over the year')
plt.show()
```

```
plt.figure(figsize = (12,3))
plt.plot(year_yield.index, year_yield['Fertilizer'],color='blue', linestyle='dashed', marker='o',
        markersize=12, markerfacecolor='green')
plt.xlabel('Year')
plt.ylabel('Fertilizer')
plt.title('Use of Fertilizer over the year')
plt.show()
```

```
plt.figure(figsize = (12,3))
plt.plot(year_yield.index, year_yield['Pesticide'],color='red', linestyle='dashed', marker='o',
        markersize=12, markerfacecolor='cyan')
plt.xlabel('Year')
plt.ylabel('Pesticide')
plt.title('Use of Pesticide over the Year')
plt.show()
```

```
data_Seas = data[data['Season']!='Whole Year ']

data_season = data_Seas.groupby('Season').sum()
data_season
```

```
fig = px.bar(data_season, y = 'Area', color=data_season.index, hover_data=['Area'],text = 'Area')
fig.show()
```

```
fig = px.sunburst(data_season, path=[data_season.index, 'Yield'], values='Yield',
                  color=data_season.index, hover_data=['Yield'])
fig.show()
```

```
data_ynz = data[data['Yield']>0]  # where yield is more than zero
data_crop = data_ynz.groupby('Crop').sum()
data_crop
```

```
plt.figure(figsize = (25,8))
plt.plot(data_crop.index, data_crop['Fertilizer'],color='red', linestyle='dashed', marker='o',
        markersize=12, markerfacecolor='cyan')
plt.xlabel('Crops')
plt.ylabel('Fertilizer')
plt.title(' Use of Fertilizer in different Crops')
plt.xticks(rotation=30)
plt.show()
```

```
plt.figure(figsize = (25,8))
plt.plot(data_crop.index, data_crop['Area'],color='indigo', linestyle='dashed', marker='o',
        markersize=12, markerfacecolor='fuchsia')
plt.xlabel('Crops')
plt.ylabel('Area under cultivation')
plt.xticks(rotation=30)
plt.show()
```

```
data_wheat = data[data['Crop']=='Wheat']
data_wheat.reset_index(drop=True,inplace=True)
data_wheat
```

```
data_wheat1 = data_wheat[data_wheat['Crop_Year']!=2020] #2020 is incomplete data
data_wheat_year = data_wheat1.groupby('Crop_Year').sum()
plt.figure(figsize = (12,5))
plt.plot(data_wheat_year.index, data_wheat_year['Yield'],color='red', linestyle='dashed', marker='o',
        markersize=12, markerfacecolor='blue')
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Yield of Wheat Crop over the Years')
plt.show()
```

```
data1 = data.copy()
data1 = data1.drop(['Crop_Year','Pesticide'], axis = 1)
```

```
category_columns = data1.select_dtypes(include = ['object']).columns
category_columns
```

```
data1 = pd.get_dummies(data1, columns = category_columns, drop_first=True)
```

```
data1.head()
```

```
#Split the data into dependent and independent variable
x =data1.drop(['Yield'], axis = 1)
y = data1['Yield']
```

```
x.head()
```

```
y.head()
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```
x_train.shape, x_test.shape, y_train.shape,y_test.shape
```

```
# to store accuracy value
accu = []
mean=[]
root_mean=[]
std=[]
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```
# Assuming you have already defined x_train, y_train, x_test, and y_test

# Initialize the Linear Regression model
lr = LinearRegression()
lr.fit(x_train, y_train)

# Predictions on training data
y_pred_train = lr.predict(x_train)

# Predictions on test data
y_pred_test = lr.predict(x_test)

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_test)
sd=np.std(y_test - y_pred_test)
print("Linear Regression Metrics:")
print(f"Accuracy: {r2:.4f}")
print(f"Mean Absolute Error: {mae:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"Standard Deviation: {sd:.4f}")
accu.append(r2)
mean.append(mae)
root_mean.append(rmse)
std.append(sd)
```

```
import matplotlib.pyplot as plt

# Assuming you have your actual values (y_test) and predicted values (y_pred)

# Scatter plot for actual vs predicted values
plt.scatter(y_test, y_pred_test)

# Add labels and title
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression: Actual vs. Predicted Values")

# Add a diagonal line for perfect prediction (optional)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')

# Show the plot
plt.grid(True)
plt.show()
```

```
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming you have X_train, y_train, X_test, y_test (features and targets)
model = DecisionTreeRegressor()
model.fit(x_train, y_train)

# Predictions
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

# Evaluation
accuracy = r2_score(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_test, squared=False))
sd= y_test.std()

print(f"Accuracy: {accuracy:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"Standard Deviation: {sd:.4f}")
accu.append(accuracy)
mean.append(mae)
root_mean.append(rmse)
std.append(sd)
```

```
import matplotlib.pyplot as plt

# Assuming you have your actual values (y_test) and predicted values (y_pred)

# Scatter plot for actual vs predicted values
plt.scatter(y_test, y_pred_test)

# Add labels and title
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Decision Tree: Actual vs. Predicted Values")

# Add a diagonal line for perfect prediction (optional)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')

# Show the plot
plt.grid(True)
plt.show()
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming you have X_train, y_train, X_test, y_test (features and targets)
model = RandomForestRegressor()
model.fit(x_train, y_train)

# Predictions
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

# Evaluation
accuracy = r2_score(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_test, squared=False))
sd = y_test.std()

print(f"Accuracy: {accuracy:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"Standard Deviation: {sd:.4f}")

accu.append(accuracy)
mean.append(mae)
root_mean.append(rmse)
std.append(sd)
```

```
import matplotlib.pyplot as plt

# Assuming you have your actual values (y_test) and predicted values (y_pred)

# Scatter plot for actual vs predicted values
plt.scatter(y_test, y_pred_test)

# Add labels and title
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Random Forest: Actual vs. Predicted Values")

# Add a diagonal line for perfect prediction (optional)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')

# Show the plot
plt.grid(True)
plt.show()
```

```
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Initialize XGBoost regressor
xg_reg = xgb.XGBRegressor(objective='reg:squarederror', colsample_bytree=0.3, learning_rate=0.1,
                          max_depth=10, alpha=10, n_estimators=10)

# Train the model
xg_reg.fit(x_train, y_train)

# Predictions on validation set
y_test_pred = xg_reg.predict(x_test)

# Calculate evaluation metrics
accuracy = xg_reg.score(x_test, y_test)  # Test accuracy
mae = mean_absolute_error(y_test, y_test_pred)  # Mean Absolute Error
rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))  # Root Mean Square Error
sd = np.std(y_test - y_test_pred)  # Standard Deviation

print(f"Accuracy: {accuracy:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"Standard Deviation: {sd:.4f}")
accu.append(accuracy)
mean.append(mae)
root_mean.append(rmse)
std.append(sd)
```

```
import matplotlib.pyplot as plt

# Assuming you have your actual values (y_test) and predicted values (y_pred)

# Scatter plot for actual vs predicted values
plt.scatter(y_test, y_test_pred)

# Add labels and title
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("XGBoost: Actual vs. Predicted Values")

# Add a diagonal line for perfect prediction (optional)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')

# Show the plot
plt.grid(True)
plt.show()
```

```
algorithm = ['LinearRegression','DecisionTreeRegressor','RandomForestRegressor','XGBoost']
accu_data = {'Accuracy':accu, 'Mean Absolute Error(MAE)':mean, 'Root Mean Square Error(RMSE)':root_mean, 'Standard Deviation(SD)':std }
model = pd.DataFrame(accu_data, index = algorithm)
model
```

```
import matplotlib.pyplot as plt

# Get the data from the DataFrame
algorithm = model.index.to_numpy()  # List of algorithm names
accu = model['Accuracy']  # accuracy series

# Create the bar plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.bar(algorithm, accu, label='Accuracy', color='g', alpha=0.7)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.title('Comparison Accuracy')
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
